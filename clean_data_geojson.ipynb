```python
# Description 0 si communce
commune_phrases = [
    'municipality in Switzerland', 'municipality in the canton of Vaud, Switzerland',
    'municipality in Vaud, Switzerland', 'municipality in the canton de Vaud, Switzerland',
    'municipality in the district of Broye-Vully in the canton of Vaud in Switzerland',
    'municipality in the canton of Vaud in Switzerland', 'municipality in the canton of Vaud inSwitzerland',
    'village in Vallée de Joux and municipality'
]

# Description 0 si village
village_phrases = [
    'former municipality', 'village in', 'locality in', 'ski resort in', 'city district of',
    'former municipality and current locality', 'village and former municipality', 'settlement in',
    'village, former municipality and current district of','Locality of the Vaud Alps in the municipality',
    'Swiss village', 'locality of the canton of Vaud', 'locality above Montreux'
]

def classify_description(description):
    if pd.isna(description):
        return description  # Garde NaN tel quel si applicable
    for phrase in commune_phrases:
        if phrase in description:
            return '0'
    for phrase in village_phrases:
        if phrase in description:
            return '1'
    return NaN

data_wikidata['Description'] = data_wikidata['Description'].apply(classify_description)

unique_descriptions = data_wikidata['Description'].unique()
print(unique_descriptions)
```
